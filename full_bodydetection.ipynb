```python
import cv2
import matplotlib.pyplot as plt
img=cv2.imread(r'D:\DL Projects\Full_body_detection\fullbody.jpg')
plt.imshow(img)
```

```python
con_img=cv2.cvtColor(img,cv2.COLOR_BGR2RGB)
plt.imshow(con_img)
```

```python
gray_img=cv2.cvtColor(img,cv2.COLOR_BGR2GRAY)
plt.imshow(gray_img)
```

```python
body_cas=cv2.CascadeClassifier(r'D:\DL Projects\Full_body_detection\haarcascade_fullbody.xml')
```

```python
fullbody=body_cas.detectMultiScale(gray_img,minNeighbors=3,scaleFactor=1.1)
```

```python
for (x,y,w,h) in fullbody:
    cv2.rectangle(con_img,(x,y),(x+w,y+h),(255,0,0),4)
```
